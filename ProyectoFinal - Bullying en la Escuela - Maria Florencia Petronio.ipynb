####**Descripción / Objetivo**
Como mamá, no puedo evitar preocuparme por mis hijos todo el tiempo, incluso cuando no están bajo mi cuidado. Desde el momento en que los dejo en la escuela por la mañana, las preocupaciones se multiplican ¿Estarán bien? ¿Se llevarán bien con sus compañeros? ¿Cómo será su día?

El objetivo principal de este proyecto es explorar el conjunto de datos para determinar los factores de riesgo asociados y las relaciones que influyen en el acoso entre estudiantes.

####**Problema comercial y contexto analítico**
En nuestro país, el acoso escolar es una realidad que afecta a muchos estudiantes. ¿Cuál es la magnitud de este problema? Bueno, según la Encuesta Mundial de Salud Escolar realizada en Argentina en 2018 -que abarcó 56,982 observaciones y 18 variables-, aproximadamente el 21% de los estudiantes encuestados han enfrentado el acoso en la escuela. Esto significa que uno de cada cinco niños ha vivido esta experiencia dolorosa. Además, identificamos que el cyber bullying y el bullying fuera de la escuela también son problemas significativos.

####**Preguntas/Problema que se busca resolver**
¿Qué factores están relacionados con el acoso escolar en Argentina? Al analizar los datos, encontramos que varios aspectos desempeñan un papel importante. Por ejemplo, la Encuesta nos muestra que los estudiantes que experimentan sentimientos de soledad tienen una mayor probabilidad de ser víctimas de bullying.
Al explorar los factores individuales, encontramos que la presencia de amigos cercanos reduce la susceptibilidad al bullying, resaltando la importancia de cuánto necesitamos promover la inclusión y la amistad en nuestras escuelas.

La implicación en peleas físicas, las faltas escolares sin permiso y la percepción de los padres acerca del problema también se relacionaron con el bullying. Los estudiantes que experimentan bullying en la escuela tienden a tener padres con un entendimiento menos profundo de la situación.

####**Contexto comercial / Audiencia**
Estos datos no son solo números, son experiencias reales de nuestros hijos. Como madres, padres, y miembros de la comunidad, tenemos la responsabilidad de tomar medidas. Debemos trabajar junto con las escuelas, educadores y autoridades para crear entornos escolares seguros, donde todos los niños se sientan protegidos y apoyados.

En última instancia, nuestros hijos merecen un futuro donde la escuela sea un lugar de aprendizaje y crecimiento, no de miedo y acoso. Estoy segura de que juntos podemos construir un camino hacia un entorno educativo en Argentina en el que todos los niños puedan prosperar y ser felices. ¡Porque como madre, nuestra preocupación nunca descansa, pero nuestra determinación también es inquebrantable!

###**3.1 Base de Datos**

Para llevar a cabo este proyecto, se utilizará el conjunto de datos de la Encuesta Mundial de Salud Escolar (GSHS) realizada en Argentina en 2018, en la cual participaron un total de 56,981 estudiantes.

###**3.1 Las variables del conjunto son las siguientes:**

*   Record: Identificador único o número de registro de cada entrada en la base de datos. Esta columna permite distinguir y rastrear individualmente cada registro.

*   Bullied_on_school_property_in_past_12_months: Indica si el estudiante fue víctima de acoso escolar en la propiedad escolar en los últimos 12 meses.

*   Bullied_not_on_school_property_in_past_12_months: Indica si el estudiante fue víctima de acoso escolar fuera de la propiedad escolar en los últimos 12 meses.

*   Cyber_bullied_in_past_12_months: Indica si el estudiante fue víctima de ciberacoso en los últimos 12 meses.

*   Custom_Age: Edad del estudiante.

*   Sex: Género o sexo del estudiante.

*   Physically_attacked: Indica si el estudiante experimentó un ataque físico en el pasado.

*   Physical_fighting: Indica si el estudiante estuvo involucrado en peleas físicas en el pasado.

*   Felt_lonely: Indica si el estudiante ha experimentado sentimientos de soledad.

*   Close_friends: Indica la cantidad de amigos cercanos que el estudiante tiene.

*   Miss_school_no_permission: Indica la cantidad de días que el estudiante faltó a la escuela sin permiso en algún momento.

*   Other_students_kind_and_helpful: Indica si otros estudiantes han sido amables y serviciales con el estudiante.

*   Parents_understand_problems: Indica si los padres o tutores del estudiante comprenden los problemas que enfrenta.

*   Most_of_the_time_or_always_felt_lonely: Indica si el estudiante se ha sentido solo la mayor parte del tiempo o siempre. Esta columna refuerza la información sobre los sentimientos persistentes de soledad del estudiante.

*   Missed_classes_or_school_without_permission: Indica si el estudiante ha faltado a clases o a la escuela sin permiso en general. Proporciona información adicional sobre la ausencia escolar no autorizada.

*   Were_underweight: Indica si el estudiante estaba por debajo del peso recomendado o considerado como peso saludable.

*   Were_overweight: Indica si el estudiante tenía sobrepeso o un índice de masa corporal superior al rango considerado saludable.

*   Were_obese: Indica si el estudiante estaba clasificado como obeso, es decir, tenía un índice de masa corporal muy por encima del rango considerado saludable.

La base de datos de análisis contiene todas variables de clasificación. Las variables de clasificación no tienen un concepto directo de "valor numérico" como las variables numéricas, por lo que la detección de outliers en este caso se basa en la frecuencia o distribución de las categorías.

####**Frecuencia de Categorías**

A continuación se realiza un análisis para identificar categorías poco comunes o inusuales en comparación con la distribución general de categorías. Este enfoque se basa en la frecuencia de las categorías y su coherencia con el contexto del problema. Si una categoría tiene una frecuencia significativamente más alta o más baja que las demás, podría considerarse como un outlier en términos de la distribución de categorías.

Aquí están las diferencias entre esas métricas:

Accuracy (Exactitud): La Accuracy se refiere a la proporción de predicciones correctas sobre el total de predicciones realizadas por el modelo.

Precisión: La Precisión es la proporción de predicciones positivas (verdaderos positivos) entre todas las instancias que el modelo ha clasificado como positivas (verdaderos positivos más falsos positivos).

Recall (Recuperación o Sensibilidad): El Recall es la proporción de instancias positivas (verdaderos positivos) que el modelo ha predicho correctamente sobre el total de instancias que son realmente positivas (verdaderos positivos más falsos negativos).

F1-score: El F1-score es una medida que combina tanto la Precisión como el Recall en una sola métrica. Es útil cuando deseas encontrar un equilibrio entre la precisión y la recuperación. Un F1-score más alto indica un equilibrio mejor entre estas dos métricas.

Matriz de Confusión: La matriz de confusión es una tabla que muestra la cantidad de predicciones correctas e incorrectas del modelo en función de las categorías reales.

####**7.4.1 KMeans**

El método de K-means es utilizado para realizar agrupamientos o segmentaciones en conjuntos de datos. Su objetivo principal es particionar un conjunto de observaciones en k grupos (clusters) de manera que cada observación pertenezca al grupo cuyo centroide (punto central) es el más cercano.

```python
# diccionarios vacios para ser almacenados los resultados
sse,db,slc,cal = {}, {}, {}, {}
for k in range(2, 10):
    # semilla de 10 para asegurar reproducibilidad
    kmeans = KMeans(n_clusters=k, max_iter=1000,random_state=10).fit(data_reducida) # Algoritmo K means
    if k == 3: labels = kmeans.labels_
    clusters = kmeans.labels_ # Extraccion de labels
    sse[k] = kmeans.inertia_ # Inercia: Suma de las distancias de las muestras a su centroide mas cercano
    db[k] = davies_bouldin_score(data_reducida,clusters) # Calculo del indice de Davies Bouldin
    slc[k] = silhouette_score(data_reducida,clusters) # Calculo del indice de Silhouette
    cal[k] = calinski_harabasz_score(data_reducida,clusters) # Calculo del indice de Calinski-Harabasz
```

####**Método de codo**

El método del codo (Elbow Method) es una técnica utilizada en el análisis de clustering, específicamente con el algoritmo K-means, para determinar el número óptimo de clusters en un conjunto de datos.
La idea detrás del método del codo es que a medida que aumentas el número de clusters (k), la WCSS (Cálculo de la Suma de Cuadrados Intra-cluster) tiende a disminuir, ya que los centroides se adaptan mejor a los datos. Sin embargo, después de cierto punto, agregar más clusters no proporciona una mejora significativa y la disminución en WCSS comienza a ser menos pronunciada, formando así el "codo" en la gráfica.

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# Crear el gráfico de líneas
plt.plot(list(sse.keys()), list(sse.values()), color='#FFD1A4')

# Configurar etiquetas
plt.xlabel("Número de clusters")
plt.ylabel("SSE")

# Mostrar el gráfico
plt.show()
```

El punto óptimo de clusters se determina comúnmente identificando el "codo" en la gráfica.

En este gráfico, no es sencillo discernir un punto de codo claro (pudiendo considerarse en n=4 o n=6). La elección del número óptimo de clusters puede necesitar una evaluación adicional basada en el conocimiento del dominio o en la aplicación de técnicas complementarias.

####**Davies-Bouldin Index**

El índice Davies-Bouldin (DB) es una métrica utilizada para evaluar la calidad de los clusters en un algoritmo de clustering, como K-means. El objetivo del índice Davies-Bouldin es cuantificar la separación y la dispersión entre los clusters, buscando clusters que sean compactos y bien separados.

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# Crear el gráfico de líneas
plt.plot(list(db.keys()), list(db.values()), color='#FFD1A4')

# Configurar etiquetas
plt.xlabel("Número de clusters")
plt.ylabel("Índice Davies-Bouldin")

# Mostrar el gráfico
plt.show()
```

El número óptimo de clusters es aquel que minimiza el índice Davies-Bouldin en el conjunto total de clusters. Cuanto menor sea el índice, mejor será la calidad de los clusters. En este contexto, valores como n=6 o n=9 podrían considerarse óptimos según esta métrica.

Un valor alto de Silhouette indica una buena estructura de clustering, mientras que un valor bajo sugiere una estructura deficiente. En este caso, el valor máximo se observa en n=3.

####**7.4.2 DBSCAN**

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) es un algoritmo de clustering que agrupa puntos de datos basándose en la densidad local. A diferencia de algoritmos de clustering más tradicionales como K-Means, DBSCAN no requiere que especifiques la cantidad de clusters de antemano. En lugar de eso, identifica áreas densas de puntos en el espacio de características, formando clusters en esas regiones.

En resumen, DBSCAN es una herramienta útil cuando se desconoce el número de clusters y se quiere identificar regiones de alta densidad en el espacio de características. Es especialmente útil para conjuntos de datos donde los clusters pueden tener formas y tamaños variados (como en nuestro caso).

```python
# Convertir los nombres de las características a cadenas
data_reducida.columns = data_reducida.columns.astype(str)

# Definir rangos de valores para eps y min_samples
eps_values = np.linspace(0.1, 1.0, 10)  # Ajustar según preferencias
min_samples_values = [3, 5, 8, 10]  # Ajustar según preferencias

best_eps = None
best_min_samples = None
best_silhouette_score = -1  # Inicializar con un valor bajo

# Realizar la doble búsqueda
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        cluster_labels = dbscan.fit_predict(data_reducida)

        # Calcular el Silhouette Score
        silhouette_avg = silhouette_score(data_reducida, cluster_labels)

        # Actualizar los mejores parámetros si se encuentra un mejor Silhouette Score
        if silhouette_avg > best_silhouette_score:
            best_silhouette_score = silhouette_avg
            best_eps = eps
            best_min_samples = min_samples

# Imprimir los mejores parámetros y el Silhouette Score correspondiente
print(f"Mejor eps: {best_eps}")
print(f"Mejor min_samples: {best_min_samples}")
print(f"Mejor Silhouette Score: {best_silhouette_score}")
```

Los resultados obtenidos indican los mejores parámetros encontrados durante la búsqueda para el modelo DBSCAN.

Mejor eps: 1.0
Este es el radio máximo de la vecindad de un punto para ser considerado parte del mismo cluster por DBSCAN. Un valor de 1.0 significa que el algoritmo considerará puntos dentro de una distancia de 1.0 como parte del mismo cluster.

Mejor min_samples: 3
Este es el número mínimo de puntos requeridos para formar un cluster. En otras palabras, un conjunto de al menos 3 puntos que estén dentro del radio eps se considerará un cluster.

Mejor Silhouette Score: -0.07080411244867948
El Silhouette Score es una métrica que cuantifica qué tan bien separados están los clusteres. El valor está en el rango de [-1, 1], donde un valor más alto indica una mejor separación. En este caso, un valor de aproximadamente -0.07 sugiere que la separación entre clusteres no es muy buena, y el valor negativo sugiere que hay superposición entre los clusters.

En resumen, los resultados indican que, según la métrica de Silhouette, los clusteres obtenidos con estos parámetros no están bien separados. Puede ser útil ajustar los parámetros o explorar otras técnicas de agrupamiento dependiendo de tus objetivos y la naturaleza de tus datos.

```python
# Mejores parámetros obtenidos
best_eps = 1.0
best_min_samples = 3

# Entrenar el modelo DBSCAN con los mejores parámetros
dbscan_optimal = DBSCAN(eps=best_eps, min_samples=best_min_samples)
cluster_labels_optimal = dbscan_optimal.fit_predict(data_reducida)

# Agregar los labels del modelo al DataFrame original
data_reducida['Cluster_optimal'] = cluster_labels_optimal

# Limitar el número de clusters a mostrar (asumiendo que los clusters están etiquetados de 0 a k)
num_clusters_to_show = 10
selected_clusters = np.random.choice(np.unique(cluster_labels_optimal), num_clusters_to_show, replace=False)

# Filtrar el DataFrame para incluir solo los puntos de los clusters seleccionados
data_reducida_subset = data_reducida[data_reducida['Cluster_optimal'].isin(selected_clusters)]

# Repetir el pairplot diferenciando por clusters y utilizando los labels del modelo
sns.pairplot(data_reducida_subset, hue='Cluster_optimal', palette='pastel')
plt.show()
```
